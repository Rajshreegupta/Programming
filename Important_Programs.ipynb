```python
# KRUSKAL'S IN AN UNDIRECTED GRAPH
#checked by my own
#O(ELogE) or O(ELogV)
class Subset:
    def __init__(self):
        parent=0
        rank = 0

class Edge:
    def __init__(self,s,d,w):
        self.src = s
        self.dest = d
        self.weight = w

class Graph:
    def __init__(self,E,V):
        self.edges = [Edge(0,0,0) for i in range(E)]
        self.v = V

def SortingKey(val):
    v = val.weight
    return v
def creategraph(V,E):
    graph = Graph(E, V)
    return graph
def find(subset, x):
    if subset[x].parent!=x:
        subset[x].parent = find(subset, subset[x].parent)
    return subset[x].parent
def union(subset, x, y):
    xset = find(subset,x)
    yset = find(subset,y)
    if subset[xset].rank>subset[yset].rank:
        subset[yset].parent = xset
    elif subset[xset].rank<subset[yset].rank:
        subset[yset].parent = xset
    else:
        subset[yset].parent = xset
        subset[yset].rank+=1
    
def kruskal(graph):
    subset = [Subset() for i in range(graph.v)]
    enum=0
    mylist=[]
    for v in range(graph.v):
        subset[v].parent = v
        subset[v].rank = 0
    for i in range(E):
        if enum==graph.v-1:
            return mylist
        x = find(subset,graph.edges[i].src)
        y = find(subset, graph.edges[i].dest)
        if x==y:
            continue
        union(subset, x, y)
        mylist.append([graph.edges[i].src, graph.edges[i].dest, graph.edges[i].weight])
        enum+=1
    return 0
V= 9
E= 14
graph = creategraph(V, E)
graph.edges[0] = Edge(0,1,4)
graph.edges[1] = Edge(1,2,8)
graph.edges[2] = Edge(2,3,7)
graph.edges[3] = Edge(3,4,9)
graph.edges[4] = Edge(4,5,10)
graph.edges[5] = Edge(5,6,2)
graph.edges[6] = Edge(6,7,1)
graph.edges[7] = Edge(7,0,8)
graph.edges[8] = Edge(1,7,11)
graph.edges[9] = Edge(2,8,2)
graph.edges[10] = Edge(8,6,6)
graph.edges[11] = Edge(8,7,7)
graph.edges[12] = Edge(2,5,4)
graph.edges[13] = Edge(3,5,14)
graph.edges.sort(key=SortingKey)
# graph.edges[4] = Edge(0,3)
# print(isCyclic(graph))
l = kruskal(graph)
if l:
    for i in l:
        print(i)
    
else:
    print("Not")
```

```python
#MINHEAP
#O(Logn)
from heapq import heappush
import math

class Minheap:
    def __init__(self):
        self.heap = []
    def parent(self,i):
        return int((i-1)/2)
    def insert(self,k):           #O(Logn)
        heappush(self.heap,k)
    def extractMin(self):         #O(Logn)
        return heappop(self.heap)
    def getMin(self):             #O(1)
        return self.heap[0]
    def decreaseKey(self,i,nval): #O(Logn)
        self.heap[i] = nval
        while(i!=0 and self.heap[self.parent(i)]>self.heap[i]):
            self.heap[i], self.heap[self.parent(i)] = self.heap[self.parent(i)], self.heap[i]
    def delete(self,i):          #O(Logn)
        self.decreaseKey(i, -math.inf)
        self.extractMin()
    
myheap = Minheap()
myheap.insert(3)
myheap.insert(2)
myheap.delete(1)
myheap.insert(15)
myheap.insert(5)
myheap.insert(4)
myheap.insert(45)
print(myheap.extractMin())
print(myheap.getMin())
myheap.decreaseKey(2,1)

print(myheap.getMin())
```

```python
#BFS IN GRAPH
#O(V+E)
from collections import deque, defaultdict 
##########
#Adv of taking deque is that it popleft() in O(1) unlike queue implementation in list which takes O(n)
# which of stack O(1) is same in both list and deque
##########
class Graph:
    def __init__(self, V):
        self.V = V
        self.graph = defaultdict(list)
    def addEdge(self,u,v):
        self.graph[u].append(v)
def bfs(g,N):
    # code here
    q = deque()
    q.append(0)
    inlist={}
    mylist = []
    mylist.append(0)
    for i in range(N):
        inlist.update({i:0})
    inlist[0] = 1
    while(len(q)>0):
        u = q.popleft()       #FOR poping from a stack write q.pop(), it will pop the last elememnt
        for ele in g[u]:
            if inlist[ele]!=1:  
                q.append(ele)
                inlist[ele]=1
                mylist.append(ele)
#             else:                If cycle also has to be detected
#                 print("Cycle detected")
    return mylist
N,E = map(int, input().split())
g = Graph(N)
edges = list(map(int,input().split()))
for i in range(0,2*E,2): # OR instead of 2*E you can also write len(edges)
    u = edges[i]
    v = edges[i+1]
    g.addEdge(u,v)
l = bfs(g.graph,N)
print(l)
```

```python
#DFS Traversal of a graph by STACK
#O(V+E)
class Graph:
    def __init__(self, V):
        self.V = V
        self.graph = defaultdict(list)
    def addEdge(self,u,v):
        self.graph[u].append(v)

def dfsUtils(g,N,visited,i,l):
    u=i
    pq=[]

    pq.append(u)
    
    while(len(pq)>0):
        u = pq.pop()
        visited[u] = 1
        l.append(u)
        for ele in g[u]:
            if visited[ele]==0:
                pq.append(ele)
                visited[ele]=1
    return l

def dfs(g,N):
    visited = [0 for i in range(N)]
    l=[]
    for i in range(N):
        if(not visited[i]):
            dfsUtils(g,N,visited,i,l)
    return l
                
N,E = map(int, input().split())
g = Graph(N)
edges = list(map(int,input().split()))
for i in range(0,2*E,2): # OR instead of 2*E you can also write len(edges)
    u = edges[i]
    v = edges[i+1]
    g.addEdge(u,v)
l = dfs(g.graph,N)
print(l)
```

```python
#DIJIKTRAS'S ALGORITHM IN AN UNDIRECTED GRAPH
#FOR DIRECTED GRAPH JUST MAKE DIFFERENCE IN addEdge() FUNCTION 
# O(ElogV)
from heapq import heappush, heappop
import math
class Minheap:
    def __init__(self):
        self.heap = []
class Graph:
    def __init__(self, V):
        self.V = V
        self.adj = [[] for i in range(V)]
    def addEdge(self,u,v,w):
        self.adj[u].append([v,w])  
        self.adj[v].append([u,w])
        
    def Dijiktras(self):
        pq = Minheap()
        src=0
        heappush(pq.heap, [0,src]) #weight of first node would be 0 
        key = [math.inf for i in range(self.V)]  #can also write float('inf')
        key[0]=0
        inDig = {}
        for i in range(self.V):
            inDig.update({i:0})
        while(len(pq.heap)>0):
            w = pq.heap[0][0]
            u = pq.heap[0][1]
            heappop(pq.heap)
            inDig[u]=1
            for ele in self.adj[u]:
                v = ele[0]
                w = ele[1]
                if inDig[v]!=1 and key[v]>key[u]+w:
                    key[v] = key[u]+w
                    heappush(pq.heap,[key[v],v])
        #Distance from source 0 to all other vertices
        print("Vertex distance from source")
        for i in range(0,V):
            print(i, key[i])

g = Graph(9)
g.addEdge(0,1,4)
g.addEdge(0, 7, 8) 
g.addEdge(1, 2, 8) 
g.addEdge(1, 7, 11) 
g.addEdge(2, 3, 7) 
g.addEdge(2, 8, 2)
g.addEdge(2, 5, 4) 
g.addEdge(3, 4, 9) 
g.addEdge(3, 5, 14) 
g.addEdge(4, 5, 10) 
g.addEdge(5, 6, 2) 
g.addEdge(6, 7, 1) 
g.addEdge(6, 8, 6) 
g.addEdge(7, 8, 7)
g.Dijiktras()
```
