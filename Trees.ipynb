```python
# Insertion in a binary search tree (RECURSION) 
# Searching
# Deletion
# Height
# Balanced tree or not (i.e., diff in the left and right height of each node should not exceed 1)
class Node:
    def __init__(self, val):
        self.left = None
        self.data = val
        self.right = None
def insertion(root, node):

    if root.data>node.data:
        if root.left==None:
            root.left = node
        else:
            insertion(root.left, node)
    else:
        if root.right==None:
            root.right = node
        else:
            insertion(root.right, node)
def inorder(root):
    if root==None:
        return 
    inorder(root.left)
    print(root.data, end=" ")
    inorder(root.right)
    
def search(root, val):
    if root==None or root.data == val :
        return root
    if root.data>val:
        return search(root.left, val)
    else:
        return search(root.right, val)
    
def minValue(node):
#     if node.left==None:
#         return node.data
#     return minValue(node.left)
    current = node
    while(current.left!=None):
        current = current.left
    return current.data
    
def deletion(root, val):
    if root==None:
        return root
    if root.data>val:
        root.left = deletion(root.left, val)
        return root
    elif root.data<val:
        root.right = deletion(root.right, val)
        return root
    else:
        if root.left==None:
            temp = root.right
            return temp
        elif root.right==None:
            temp = root.left
            return temp
        else:
            min_val = minValue(root.right)
            root.data = min_val
            root.right = deletion(root.right, min_val)
            return root
        
def height(root):
    if root==None:
        return 0
    return 1+max(height(root.left), height(root.right))
            
def isBalanced(root):
    if root==None:
        return True
    hl = height(root.left)
    hr = height(root.right)
    if abs(hl-hr)<=1 and isBalanced(root.left) and isBalanced(root.right):
        return True
    return False
            
N = int(input())
l = list(map(int, input().split()))
root = Node(l[0])
for i in range(1,N):
    node = Node(l[i])
    insertion(root, node)
inorder(root)

# SEARCH A KEY
val = int(input())
node = search(root, val)
if node==None:
    print("Not found")
else:
    print("Found")
    
# DELETE A KEY
val = int(input())
root = deletion(root, val)
inorder(root)

# BALANCED TREE OR NOT
print(isBalanced(root))
```

AVL TREES
AVL tree is a self-balancing Binary Search Tree (BST) 
**where the difference between heights of left and right subtrees cannot be more than 1 for all nodes.

**Why AVL Trees?**
Most of the BST operations (e.g., search, max, min, insert, delete.. etc) take O(h) time where h is the height of the BST. 
The cost of these operations may become O(n) for a skewed Binary tree. If we make sure that height of the tree remains O(Logn) after every insertion and deletion, then we can guarantee an upper bound of **O(Logn) for all these operations. The height of an AVL tree is always O(Logn) where n is the number of nodes in the tree**
